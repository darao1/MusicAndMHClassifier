```python
#major_category in this instance is not the same as above because I am defining a functio
#no indexing because my pop_feature is a string?
def create_features(major_category):
    return {'pop_feature':major_category}
```

```python
# combining my features column with my lables column (a tuple)
features_and_labels = zip(results.Major_category, results.Major)
features_and_labels
```

```python
featuresets = [(create_features(major_category), major) for major_category, major in features_and_labels]
featuresets
```
