```python
#defining my age feature
def create_features(ages):
    return {'participant_age':ages}
```

```python
# combining my features column with my labels column (a tuple)
features_and_labels = zip(results.Age, results.stream_service)
features_and_labels
```

```python
#obtaining the two column data in the format specified
featuresets = [(create_features(ages), stream_service) for ages, stream_service in features_and_labels]
featuresets
```

```python
#setting up my training data and testing data
import random
random.shuffle(featuresets)
```

```python
#there are 736 items within my featuresets list
#i.e. 736 tuples of ages and streaming services
len(featuresets)
```

```python
#approx. 80% of a dataset should be used in a training set
736*.8
```

```python
training_set = featuresets[:588]
testing_set = featuresets[588:]
```

```python
training_set
```

```python
#Building Classifier
import nltk
classifier = nltk.NaiveBayesClassifier.train(training_set)
```

```python
#testing the classsifier
classifier.classify(create_features('18.0'))
```

```python
#obtaining classifiers accuracy
print(nltk.classify.accuracy(classifier,testing_set))
```

```python
#what were the most informative features?
classifier.show_most_informative_features(5)
```
